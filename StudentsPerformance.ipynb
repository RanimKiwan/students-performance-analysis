In this project, we analyze the dataset "StudentsPerformance.csv" to explore the relationship between demographic factors and students' performance in math, reading, and writing.

```python
#1-Impoerting the main libraries:
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Loading the dataset
We load the dataset "StudentsPerformance.csv" into a pandas DataFrame 
to start exploring and analyzing the data.

```python
#2-Loading the dataset into a DataFrame:

df = pd.read_csv(r"D:\Ranim\Data_Projects\StudentsPerformance.csv",sep=';')
df
```

## Step 3: Initial data exploration
We check:

# Step 4: Exploratory Data Analysis (EDA):
In this step, we perform Exploratory Data Analysis (EDA) to better understand the dataset.
We use visualizations (plots) to observe patterns, distributions, and relationships between variables.

# 4.1 Distribution of scores:
We plot histograms to visualize the distribution of students' scores in math, reading, and writing.

```python
# 4.1 Distribution of scores:
plt.figure(figsize=(15,5))   # create figure with size (15x5)

plt.subplot(1,3,1)   # first plot (row=1, col=3, position=1)
sns.histplot(df['math score'], bins=20, kde=True, color='blue')   # histogram for math scores
plt.title("Math Score Distribution")   # title for first plot

plt.subplot(1,3,2)   # second plot (row=1, col=3, position=2)
sns.histplot(df['reading score'], bins=20, kde=True, color='green')   # histogram for reading scores
plt.title("Reading Score Distribution")   # title for second plot

plt.subplot(1,3,3)   # third plot (row=1, col=3, position=3)
sns.histplot(df['writing score'], bins=20, kde=True, color='red')   # histogram for writing scores
plt.title("Writing Score Distribution")   # title for third plot

plt.show()   # display all plots
```

# Observations after distribution plots:

-The math scores are roughly centered around 65â€“70, with fewer students at the very low and very high  ends.

-Reading scores show a similar distribution, slightly shifted towards higher values.

-Writing scores are also similar to reading, but the distribution seems a bit more skewed towards the  higher end.

-Overall, most students cluster around the mid to high performance range.

# 4.2 Comparing Scores by Gender

We use boxplots to compare the performance of male and female students in math, reading, and writing.

```python
# 4.2 Comparing scores by gender:

plt.figure(figsize=(15,5))   # create a figure with width=15 and height=5

plt.subplot(1,3,1)   # first plot in a row of 3 plots
sns.boxplot(x='gender', y='math score', data=df)   # boxplot for math scores by gender
plt.title("Math Score by Gender")   # set title for first plot

plt.subplot(1,3,2)   # second plot in the row
sns.boxplot(x='gender', y='reading score', data=df)   # boxplot for reading scores by gender
plt.title("Reading Score by Gender")   # set title for second plot

plt.subplot(1,3,3)   # third plot in the row
sns.boxplot(x='gender', y='writing score', data=df)   # boxplot for writing scores by gender
plt.title("Writing Score by Gender")   # set title for third plot

plt.show()   # display all the plots
```

# Observations:

-Females tend to score slightly higher in reading and writing compared to males.

-Males perform a bit better in math compared to females.

-The distributions show some outliers (very low scores) in all three subjects.

-Overall, the differences are visible but not extremely large.

# 4.3 Relationship Between Math and Reading Scores:
We create a scatter plot to explore the relationship between math scores and reading scores.

```python
# 4.3 Relationship Between Math and Reading Scores:

plt.figure(figsize=(7,5))   # create a figure with width=7 and height=5
sns.scatterplot(x='math score', y='reading score', data=df, hue='gender')  # scatter plot of math vs reading scores, colored by gender
plt.title("Math vs Reading Scores")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:

-There is a positive relationship between math and reading scores (students who do well in math often  do well in reading).

-Both genders follow a similar trend, but females are slightly stronger in reading.

# 4.4 Relationship Between Math and Writing Scores:

We create a scatter plot to study the relationship between math scores and writing scores, and check if high performance in math is also linked to high performance in writing.

```python
# 4.4 Relationship Between Math and Writing Scores:

plt.figure(figsize=(7,5))   # create a figure with width=7 and height=5
sns.scatterplot(x='math score', y='writing score', data=df, hue='gender')  # scatter plot of math vs writing scores, colored by gender
plt.title("Math vs Writing Scores")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:

-Strong positive correlation: students with higher math scores usually have higher writing scores.

-Both genders follow a similar trend, but females are slightly stronger in reading.

# 4.5 Relationship Between Reading and Writing Scores:

We create a scatter plot to explore the relationship between reading scores and writing scores.
This helps us see if students who perform well in reading also perform well in writing.

```python
# 4.5 Relationship Between Reading and Writing Scores:

plt.figure(figsize=(7,5))   # create a figure with width=7 and height=5
sns.scatterplot(x='reading score', y='writing score', data=df, hue='gender')  # scatter plot of reading vs writing scores, colored by gender
plt.title("Reading vs Writing Scores")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:

-There is a very strong positive correlation: students with high reading scores also have high writing scores.

-This relationship is stronger compared to math vs reading or math vs writing.

-Both genders follow the same trend, almost no difference.

# Step5: Multivariate Analysis:

In this step, we analyze how other factors (besides gender) influence studentsâ€™ performance.
We will look at:

1-Parental level of education (education level of parents).
2-Lunch type (standard vs free/reduced).
3-Test preparation course (completed vs none).

By comparing these factors with studentsâ€™ scores, we can understand which external conditions have the most impact on performance.

# 5.1 Effect of Parental Education Level:

We use a boxplot to study how the education level of parents affects students' math scores.

```python
# 5.1 Effect of Parental Education Level:

plt.figure(figsize=(10,6))   # create a figure with width=10 and height=6
sns.boxplot(x='parental level of education', y='math score', data=df)   # boxplot of math scores grouped by parents' education level
plt.xticks(rotation=45)   # rotate x-axis labels by 45 degrees for readability
plt.title("Math Scores by Parental Education Level")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:
Students whose parents have higher education (college or masterâ€™s degree) generally score higher in math.

Students whose parents have lower education levels tend to have more variation and lower average scores.

# 5.2 Effect of Lunch Type:
We want to see if the type of lunch (standard vs. free/reduced) has an impact on studentsâ€™ math scores.

```python
# 5.2 Effect of Lunch Type:

plt.figure(figsize=(6,5))   # create a figure with width=6 and height=5
sns.boxplot(x='lunch', y='math score', data=df)   # boxplot of math scores grouped by lunch type
plt.title("Math Scores by Lunch Type")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:

-Students with standard lunch perform better in math compared to students with free/reduced lunch.

-Free/reduced lunch students show lower averages and more variability.

# 5.3 Effect of Test Preparation Course:
We want to analyze whether completing a test preparation course improves studentsâ€™ math performance.

```python
# 5.3 Effect of Test Preparation Course:

plt.figure(figsize=(6,5))   # create a figure with width=6 and height=5
sns.boxplot(x='test preparation course', y='math score', data=df)   # boxplot of math scores grouped by test preparation course
plt.title("Math Scores by Test Preparation Course")   # set the title of the plot
plt.show()   # display the plot
```

# Observations:

-Students who completed the test preparation course scored higher in math compared to those who did not.

-Students without preparation show lower averages and more variability.

# Step6: Correlation Analysis:

We calculate the correlation between all numeric variables (math, reading, writing scores).
Then we visualize the correlation matrix using a heatmap to identify the strongest relationships.

```python
# calculate correlation matrix
corr = df[['math score', 'reading score', 'writing score']].corr()   # correlation between scores

# display correlation matrix
print(corr)   # print correlation values

# plot heatmap
plt.figure(figsize=(6,4))   # create figure
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")   # heatmap with values shown
plt.title("Correlation between Scores")   # set title
plt.show()   # display plot
```

# Observations:

-Strong correlation between reading and writing scores (close to 0.95).

-Math also correlates positively with reading and writing, but less strong (~0.8).

-This confirms that students good at one subject often perform well in the others.

# Step7: Conclusion:

-Most studentsâ€™ scores are concentrated in the mid to high range.

-Females perform slightly better in reading and writing, while males are a bit stronger in math.

-Strong positive correlations exist between all subjects, especially between reading and writing.

-Parentsâ€™ education level has a clear impact: students with more educated parents perform better.

-Lunch type also matters: students with standard lunch achieve higher scores.

-Completing a test preparation course significantly improves performance.
